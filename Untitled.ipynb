```python
import csv
```

```python
f = open('result.csv', 'r')
```

```python
rdr = csv.reader(f)
```

```python
result = []
for line in rdr :
    result.append(line)
```

```python
res_int = []
res_text = []
for re in result:
    res_int.append([int(i) for i in re[:-1]])
    res_text.append(re[-1])
```

```python
res_text
```

```python
res_text = [i.lower() for i in res_text]
```

```python
res_text
```

```python
import numpy as np
```

```python
result_key = np.array(res_int)
result_value = np.array(res_text)
```

```python
np.save('./result_key', result_key)
np.save('./result_value', result_value)
```
